```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from scipy.spatial import distance
from scipy.optimize import fsolve
```

```python
## Oscilloscope
df_osc = pd.read_csv('./ass2_data/FS.csv')

time_osc = df_osc['in s']
clk_osc  = df_osc['C1 in V']
spk_osc  = df_osc['C2 in V']

fig, axs = plt.subplots(2, 1, figsize=(8, 4))
axs[0].plot(time_osc, clk_osc, color='black', alpha=0.2, label='CLK')
axs[0].plot(time_osc, spk_osc, color='black', linewidth=2, label='SPIKE')

## Simulation
df_sim = pd.read_csv('./ass2_data/FS_sim.csv')
df_sim = df_sim.replace('U', '0')

clk_sim  = df_sim['clk']
spk_sim  = df_sim['spike']

clk_freq    = 1000e3
clk_period  = 1 / clk_freq

time_sim = np.arange(len(clk_sim)) * clk_period

axs[1].step(time_sim, clk_sim, where='pre', color='black', linewidth=2, alpha=0.2, label='CLK')
axs[1].step(time_sim, spk_sim, where='post', color='black', linewidth=2, label='SPIKE')

axs[0].set_xlim(1e-6, 30e-6)
axs[1].set_xlim(41e-6, 99e-6)

axs[0].set_yticks([0, 3.3])
axs[1].set_yticks([0, 1])

axs[0].set_yticklabels(['0 V', '3.3 V'])

axs[0].legend(loc='upper right', framealpha=1)
axs[1].legend(loc='upper right', framealpha=1)

axs[0].hlines(y=3.3/2, xmin=0.41e-5, xmax=1.238e-5, colors='blue', linestyles='--', linewidth=3)
axs[0].text(0.75e-5, 3.3/2+3.3/10, f'8 cycles', fontsize=12,
            bbox=dict(facecolor='white', edgecolor='blue', boxstyle='round,pad=0.3'))

axs[1].hlines(y=0.5, xmin=4.8e-5, xmax=6.4e-5, colors='blue', linestyles='--', linewidth=3)
axs[1].text(5.5e-5, 0.6, f'8 cycles', fontsize=12,
            bbox=dict(facecolor='white', edgecolor='blue', boxstyle='round,pad=0.3'))

axs[0].text(0.02, 0.92, '(a)', transform=axs[0].transAxes, fontsize=14, fontweight='bold', 
            va='top', ha='left', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
axs[1].text(0.02, 0.92, '(b)', transform=axs[1].transAxes, fontsize=14, fontweight='bold', 
            va='top', ha='left', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

axs[0].get_xaxis().set_visible(False)
axs[1].get_xaxis().set_visible(False)
fig.tight_layout()

fig.savefig(f'../assignment_2/report/images/fs_sim_v_fpga.svg')
```
